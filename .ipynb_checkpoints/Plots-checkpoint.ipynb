```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
sorted_movieId = pd.read_csv('processed_data/sorted_movieId.csv', header = None)
```

```python
sns.set(style='white')
fig, ax = plt.subplots()
ax.plot(range(sorted_movieId.shape[0]),sorted_movieId[1])
ax.set_xlabel('Rank')
ax.set_ylabel('Popularity')
plt.savefig('plots/long-tail.pdf', bbox_inches = 'tight')
plt.show()
```

```python
GANLAMBDA = [0.1, 1.0, 10.0, 100.0]
epoch = 80
```

```python
style = ['b-','r','g-','m-']
fig, ax = plt.subplots(figsize=(8,4))
for i in range(4):
    scores_log_path = 'long-tail_GAN/train_log/ml_1m_holdout_user_LT_GAN_{}_0.0001/scores_log/scores_79.txt'.format(GANLAMBDA[i])
    scores_log = np.loadtxt(scores_log_path,delimiter = ',',skiprows=1)
    ndcg = scores_log[:,2]
    ax.plot(ndcg,style[i],label='lambda = {}'.format(GANLAMBDA[i]))
    ax.set_xlabel('Global Epochs')
    ax.set_ylabel('NDCG@100')
ax.legend(loc=4)
plt.savefig('plots/long-tail GAN(training).pdf', bbox_inches = 'tight')
plt.show()
```
